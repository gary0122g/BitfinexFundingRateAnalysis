```python
import requests
import pandas as pd
import numpy as np
import time
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Retain the original data acquisition function
def get_fusd_funding_stats(start=None, end=None, limit=1000):
    """
    Get Bitfinex fUSD funding statistics data

    Parameters:
    start (int): Start timestamp (milliseconds), optional
    end (int): End timestamp (milliseconds), optional
    limit (int): Maximum number of records to return, max value is 250, default is 250

    Returns:
    pandas DataFrame containing fUSD funding statistics data
    """
    url = "https://api-pub.bitfinex.com/v2/funding/stats/fUSD/hist"

    # Build request parameters
    params = {"limit": limit}
    if start:
        params["start"] = start
    if end:
        params["end"] = end

    try:
        response = requests.get(url, params=params)
        if response.status_code == 200:
            data = response.json()

            # Define column names according to API documentation
            columns = [
                'mts',           # Millisecond timestamp
                'placeholder1',  # Unused
                'placeholder2',  # Unused
                'frr',           # Flash Return Rate (1/365)
                'avg_period',    # Average funding period
                'placeholder3',  # Unused
                'placeholder4',  # Unused
                'funding_amount',        # Total funding amount provided
                'funding_amount_used',   # Funding amount used
                'placeholder5',  # Unused
                'placeholder6',  # Unused
                'funding_below_threshold' # Total funding below threshold
            ]

            df = pd.DataFrame(data, columns=columns)

            # Convert timestamp to readable format
            df['datetime'] = pd.to_datetime(df['mts'], unit='ms')

            # Calculate daily rate (%)
            df['daily_rate_percent'] = df['frr'] * 365 * 100

            # Calculate annualized rate (APR %)
            df['apr_percent'] = df['frr'] * 365 * 365 * 100

            # Calculate usage ratio (%)
            df['usage_ratio'] = df['funding_amount_used'] / df['funding_amount'] * 100

            # Keep only useful columns
            useful_columns = [
                'mts', 'datetime', 'frr', 'daily_rate_percent', 'apr_percent',
                'avg_period', 'funding_amount', 'funding_amount_used', 'usage_ratio',
                'funding_below_threshold'
            ]

            return df[useful_columns]
        else:
            print(f"Error: {response.status_code}, {response.text}")
            return None
    except Exception as e:
        print(f"Error getting fUSD funding statistics data: {e}")
        return None

def get_btc_price_data(timeframe="1m", start=None, end=None, limit=1000):
    """
    Get Bitfinex BTC/USD price data

    Parameters:
    timeframe (str): Time frame, e.g. "1m", "5m", "15m", "1h", "1D", etc.
    start (int): Start timestamp (milliseconds), optional
    end (int): End timestamp (milliseconds), optional
    limit (int): Maximum number of records to return, max value is 1000, default is 250

    Returns:
    pandas DataFrame containing BTC/USD price data
    """
    # Use URL encoded colon (%3A)
    url = f"https://api-pub.bitfinex.com/v2/candles/trade%3A{timeframe}%3AtBTCUSD/hist"

    # Build request parameters
    params = {"limit": limit}

    # Add request headers
    headers = {"accept": "application/json"}

    try:
        print(f"Requesting URL: {url} Parameters: {params}")
        response = requests.get(url, params=params, headers=headers)

        if response.status_code == 200:
            data = response.json()

            # Check if valid data was returned
            if isinstance(data, list) and len(data) > 0:
                # Define column names according to API documentation
                columns = [
                    'mts',    # Millisecond timestamp
                    'open',   # Open price
                    'close',  # Close price
                    'high',   # High price
                    'low',    # Low price
                    'volume'  # Trading volume
                ]

                df = pd.DataFrame(data, columns=columns)

                # Convert timestamp to readable format
                df['datetime'] = pd.to_datetime(df['mts'], unit='ms')

                # Calculate price change percentage
                df['price_change_percent'] = ((df['close'] - df['open']) / df['open']) * 100

                # Calculate volatility (percentage difference between high and low prices)
                df['volatility_percent'] = ((df['high'] - df['low']) / df['low']) * 100

                return df
            else:
                print(f"API returned empty data or error: {data}")
                return None
        else:
            print(f"Error getting BTC price data: {response.status_code}, {response.text}")
            return None
    except Exception as e:
        print(f"Error getting BTC price data: {e}")
        return None

def save_to_csv(data, filename="bitfinex_fusd_funding_stats.csv"):
    """
    Save data to CSV file
    """
    if data is not None:
        data.to_csv(filename, index=False)
        print(f"Data saved to {filename}")

def plot_funding_rates(data, filename="fusd_funding_rate.png"):
    """
    Plot fUSD funding rate trend (English version)
    """
    if data is not None and not data.empty:
        plt.figure(figsize=(14, 8))
        plt.plot(data['datetime'], data['daily_rate_percent'])
        plt.title('Bitfinex fUSD Daily Funding Rate Trend (%)')
        plt.xlabel('Date')
        plt.ylabel('Daily Rate (%)')
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(filename)
        plt.show()

def get_historical_fusd_data(days=90):  # Increase default days to get more training data
    """
    Get historical fUSD funding rate data for specified number of days
    Due to API limitations returning max 250 records per request, we need to fetch in batches

    Parameters:
    days (int): Number of days to fetch

    Returns:
    pandas DataFrame containing all historical data
    """
    end_time = int(time.time() * 1000)  # Current time (milliseconds)

    # Create an empty DataFrame to store all data
    all_data = pd.DataFrame()

    # Estimate the number of batches needed (assuming one record every 3 hours)
    batches = max(1, int(days * 24 / 3 / 250) + 1)

    for i in range(batches):
        if i > 0:
            # Update end time to be the earliest record time minus 1 millisecond
            end_time = all_data['mts'].min() - 1

        print(f"Fetching batch {i+1}/{batches}, end time: {datetime.fromtimestamp(end_time/1000)}")

        batch_data = get_fusd_funding_stats(end=end_time, limit=250)

        if batch_data is not None and not batch_data.empty:
            all_data = pd.concat([all_data, batch_data])

            # Check if we've fetched enough data
            oldest_date = all_data['datetime'].min()
            newest_date = all_data['datetime'].max()
            date_range = (newest_date - oldest_date).days

            print(f"  Fetched data from {oldest_date} to {newest_date} ({date_range} days)")

            if date_range >= days:
                print(f"  Reached target days {days}, stopping fetch")
                break
        else:
            print("  Could not fetch more data, stopping")
            break

        time.sleep(1)  # Avoid API rate limits

    # Sort and remove duplicates
    if not all_data.empty:
        all_data = all_data.sort_values('mts').drop_duplicates(subset=['mts'])
        print(f"Total of {len(all_data)} fUSD funding data records fetched")

    return all_data

# New function: Merge fUSD data and BTC price data
def merge_fusd_and_btc_data(fusd_data, btc_data):
    """
    Merge fUSD funding data and BTC price data

    Parameters:
    fusd_data: fUSD funding data DataFrame
    btc_data: BTC price data DataFrame

    Returns:
    Merged DataFrame
    """
    if fusd_data is None or btc_data is None or fusd_data.empty or btc_data.empty:
        print("Cannot merge data: fUSD data or BTC price data is empty")
        return fusd_data

    # Ensure data is sorted by time
    fusd_data = fusd_data.sort_values('datetime')
    btc_data = btc_data.sort_values('datetime')

    # Reindex BTC data to fUSD data time points
    # Use forward fill method, i.e., for each fUSD data point, use the most recent past BTC data
    merged_data = pd.merge_asof(
        fusd_data,
        btc_data[['datetime', 'close', 'price_change_percent', 'volatility_percent', 'volume']],
        on='datetime',
        direction='backward'
    )

    # Rename columns to avoid confusion
    merged_data = merged_data.rename(columns={
        'close': 'btc_price',
        'price_change_percent': 'btc_price_change_percent',
        'volatility_percent': 'btc_volatility_percent',
        'volume': 'btc_volume'
    })

    # Fill possible missing values
    merged_data = merged_data.fillna(method='ffill')

    print(f"Merged data contains {len(merged_data)} rows and the following columns:")
    print(merged_data.columns.tolist())

    return merged_data

# Multi-feature LSTM prediction related functions
def create_multifeature_sequences(data, seq_length, feature_columns):
    """
    Create sequence data for multi-feature LSTM

    Parameters:
    data: Original data DataFrame
    seq_length: Sequence length (how many past data points to use to predict the next point)
    feature_columns: List of feature column names to use

    Returns:
    X: Input sequences
    y: Target values
    """
    X, y = [], []
    for i in range(len(data) - seq_length):
        # Take sequences of multiple features
        X.append(data[i:i + seq_length, :])
        # Only predict daily rate
        y.append(data[i + seq_length, 0])  # 0 is the index of daily rate in features
    return np.array(X), np.array(y)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input

def build_multifeature_lstm_model(seq_length, n_features, units=50):
    """
    Build multi-feature LSTM model - corrected Sequential API
    """
    model = Sequential([
        # Use Input layer as the first layer
        Input(shape=(seq_length, n_features)),

        LSTM(units=units, return_sequences=True),
        Dropout(0.5),

        LSTM(units=units),  # Last LSTM layer doesn't return sequences
        Dropout(0.2),

        Dense(1)
    ])

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def train_multifeature_lstm_model(data, feature_columns, seq_length=5, epochs=100, batch_size=32, validation_split=0.3):
    """
    Train multi-feature LSTM model to predict funding rate

    Parameters:
    data: DataFrame containing multiple features
    feature_columns: List of feature column names to use
    seq_length: Sequence length
    epochs: Number of training epochs
    batch_size: Batch size
    validation_split: Validation set proportion

    Returns:
    model: Trained model
    scalers: Dictionary of scalers for denormalization
    X_test: Test set inputs
    y_test: Test set targets
    feature_data: Normalized feature data
    """
    # Ensure data is sorted by time
    data = data.sort_values('datetime').reset_index(drop=True)

    # Create a dictionary to store scalers for each feature
    scalers = {}
    scaled_features = {}

    # Create and apply scaler for each feature
    for feature in feature_columns:
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_features[feature] = scaler.fit_transform(data[feature].values.reshape(-1, 1)).flatten()
        scalers[feature] = scaler

    # Combine all normalized features into an array
    feature_data = np.column_stack([scaled_features[feature] for feature in feature_columns])

    # Create sequence data
    X, y = create_multifeature_sequences(feature_data, seq_length, feature_columns)

    # Split into training and test sets (use last 20% of data as test set)
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Build model
    model = build_multifeature_lstm_model(seq_length, len(feature_columns))

    # Set up early stopping to avoid overfitting
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True
    )

    # Train model
    print("Starting multi-feature LSTM model training...")
    print(f"Using features: {feature_columns}")
    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_split=validation_split,
        callbacks=[early_stopping],
        verbose=1
    )

    # Plot training process
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Multi-feature LSTM Model Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.savefig('multifeature_lstm_training_history.png')
    plt.show()

    return model, scalers, X_test, y_test, X_train, y_train, feature_data

def evaluate_multifeature_model(model, X_test, y_test, scalers, target_feature):
    """
    Evaluate multi-feature model performance

    Parameters:
    model: Trained model
    X_test: Test set inputs
    y_test: Test set targets
    scalers: Dictionary of scalers for denormalization
    target_feature: Target feature name

    Returns:
    Evaluation metrics
    """
    # Predict
    y_pred = model.predict(X_test)

    # Denormalize
    target_scaler = scalers[target_feature]
    y_test_original = target_scaler.inverse_transform(y_test.reshape(-1, 1))
    y_pred_original = target_scaler.inverse_transform(y_pred)

    # Calculate evaluation metrics
    mse = mean_squared_error(y_test_original, y_pred_original)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test_original, y_pred_original)
    r2 = r2_score(y_test_original, y_pred_original)

    print(f"Mean Squared Error (MSE): {mse:.6f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.6f}")
    print(f"Mean Absolute Error (MAE): {mae:.6f}")
    print(f"Coefficient of Determination (R²): {r2:.6f}")

    return mse, rmse, mae, r2

def plot_multifeature_predictions(data, model, scalers, feature_columns, target_feature, seq_length, future_days=7):
    """
    Plot multi-feature prediction results and predict funding rates for future days

    Parameters:
    data: Original data
    model: Trained model
    scalers: Dictionary of scalers for normalization/denormalization
    feature_columns: List of features used
    target_feature: Target feature name
    seq_length: Sequence length
    future_days: Number of future days to predict
    """
    # Ensure data is sorted by time
    data = data.sort_values('datetime').reset_index(drop=True)

    # Normalize all features
    scaled_features = {}
    for feature in feature_columns:
        scaled_features[feature] = scalers[feature].transform(data[feature].values.reshape(-1, 1)).flatten()

    # Combine all normalized features into an array
    feature_data = np.column_stack([scaled_features[feature] for feature in feature_columns])

    # Predict historical data
    X = []
    for i in range(len(feature_data) - seq_length):
        X.append(feature_data[i:i + seq_length])
    X = np.array(X)

    predictions = model.predict(X)
    target_scaler = scalers[target_feature]
    predictions = target_scaler.inverse_transform(predictions)

    # Create time axis for predictions
    pred_dates = data['datetime'].iloc[seq_length:].reset_index(drop=True)

    # Plot historical data and predictions
    plt.figure(figsize=(16, 8))
    plt.plot(data['datetime'], data[target_feature], label='Actual Rate')
    plt.plot(pred_dates, predictions, label='Model Prediction', color='red', linestyle='--')

    # Predict future data
    future_predictions = []
    last_sequence = feature_data[-seq_length:].copy()

    future_dates = []
    last_date = data['datetime'].iloc[-1]

    # Assume data interval is 3 hours
    time_interval = 3

    # For future prediction, we need to estimate future values of other features
    # Here we use a simple method: assume other features remain at their last known value
    last_values = {}
    for i, feature in enumerate(feature_columns):
        if feature != target_feature:  # For non-target features
            last_values[i] = last_sequence[-1, i]

    for i in range(future_days * 24 // time_interval):  # 24/time_interval data points per day
        # Use model to predict next point
        next_pred = model.predict(last_sequence.reshape(1, seq_length, len(feature_columns)))

        # Add to prediction list
        future_predictions.append(next_pred[0, 0])

        # Create feature vector for next time point
        next_features = np.zeros(len(feature_columns))
        next_features[0] = next_pred[0, 0]  # Target feature (daily rate) uses predicted value

        # Other features use last known value
        for i, feature in enumerate(feature_columns):
            if feature != target_feature:  # For non-target features
                next_features[i] = last_values[i]

        # Update sequence (remove first element, add prediction)
        last_sequence = np.vstack([last_sequence[1:], next_features])

        # Calculate next time point
        next_date = last_date + timedelta(hours=time_interval)
        future_dates.append(next_date)
        last_date = next_date

    # Denormalize future predictions
    future_predictions = np.array(future_predictions).reshape(-1, 1)
    future_predictions = target_scaler.inverse_transform(future_predictions)

    # Plot future predictions
    plt.plot(future_dates, future_predictions, label='Future Forecast', color='green', linestyle='-.')

    # Add vertical line marking current time
    plt.axvline(x=data['datetime'].iloc[-1], color='purple', linestyle='--', alpha=0.7)
    plt.text(data['datetime'].iloc[-1], plt.ylim()[1]*0.9, 'Current', rotation=90, color='purple')

    plt.title('Bitfinex fUSD Daily Funding Rate - Multi-feature Prediction')
    plt.xlabel('Date')
    plt.ylabel('Daily Rate (%)')
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig('fusd_multifeature_prediction.png')
    plt.show()

    # Output future prediction values
    print("\nFuture fUSD Funding Rate Predictions (Multi-feature Model):")
    future_df = pd.DataFrame({
        'datetime': future_dates,
        'predicted_daily_rate': future_predictions.flatten()
    })

    # Group by day and calculate mean (since we may have multiple predictions per day)
    future_df['date'] = future_df['datetime'].dt.date
    daily_predictions = future_df.groupby('date')['predicted_daily_rate'].mean().reset_index()

    for _, row in daily_predictions.iterrows():
        print(f"{row['date']}: {row['predicted_daily_rate']:.4f}%")

    return future_df

def plot_feature_importance(data, feature_columns):
    """
    Analyze and plot feature correlations

    Parameters:
    data: Data DataFrame
    feature_columns: List of features
    """
    # Calculate correlation matrix
    correlation = data[feature_columns].corr()

    # Plot correlation heatmap
    plt.figure(figsize=(12, 10))
    plt.title('Feature Correlation Matrix')

    # Use heatmap to display correlations
    im = plt.imshow(correlation, cmap='coolwarm')
    plt.colorbar(im)

    # Add correlation value labels
    for i in range(len(correlation)):
        for j in range(len(correlation)):
            text = plt.text(j, i, f'{correlation.iloc[i, j]:.2f}',
                           ha="center", va="center", color="black")

    # Set axis labels
    plt.xticks(range(len(correlation)), correlation.columns, rotation=90)
    plt.yticks(range(len(correlation)), correlation.columns)

    plt.tight_layout()
    plt.savefig('feature_correlation.png')
    plt.show()

    # Correlation with target variable
    target = 'daily_rate_percent'
    target_corr = correlation[target].drop(target).sort_values(ascending=False)

    plt.figure(figsize=(10, 6))
    plt.title(f'Feature Correlation with {target}')
    target_corr.plot(kind='bar')
    plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.savefig('target_correlation.png')
    plt.show()

    print("\nFeature correlations with daily rate:")
    for feature, corr in target_corr.items():
        print(f"{feature}: {corr:.4f}")

def main():
    print("Starting to fetch Bitfinex fUSD funding rate data and BTC price data...")

    # 1. Get latest fUSD funding rate
    latest_data = get_fusd_funding_stats(limit=10)
    if latest_data is not None:
        print("\nLatest fUSD funding rates:")
        print(latest_data[['datetime', 'daily_rate_percent', 'apr_percent']].head())

        # Display latest rate
        latest_rate = latest_data['daily_rate_percent'].iloc[0]
        latest_apr = latest_data['apr_percent'].iloc[0]
        print(f"\nCurrent fUSD daily rate: {latest_rate:.4f}%")
        print(f"Current fUSD annualized rate (APR): {latest_apr:.4f}%")

    # 2. Get historical fUSD data (increased to 90 days to get more training data)
    print("\nStarting to fetch historical fUSD funding data...")
    historical_fusd_data = get_historical_fusd_data(days=90)

    # 3. Get historical BTC price data
    print("\nStarting to fetch historical BTC price data...")
    start_time = None
    end_time = None

    if not historical_fusd_data.empty:
        # Use fUSD data time range to get BTC data
        start_time = historical_fusd_data['mts'].min()
        end_time = historical_fusd_data['mts'].max() + 24*60*60*1000  # Get one more day of data

    historical_btc_data = get_btc_price_data(timeframe="1h", start=start_time, end=end_time, limit=1000)

    if historical_btc_data is not None:
        print(f"Fetched {len(historical_btc_data)} BTC price data records")

    # 4. Merge fUSD and BTC data
    merged_data = merge_fusd_and_btc_data(historical_fusd_data, historical_btc_data)

    # 5. Save merged data to CSV
    if merged_data is not None and not merged_data.empty:
        today = datetime.now().strftime('%Y%m%d')
        save_to_csv(merged_data, f"bitfinex_fusd_btc_merged_data_{today}.csv")

    # 6. Plot trend chart
    plot_funding_rates(historical_fusd_data)

    # 7. Plot BTC price and fUSD rate comparison chart
    if merged_data is not None and not merged_data.empty:
        plt.figure(figsize=(16, 12))

        # Create two subplots
        ax1 = plt.subplot(211)
        ax2 = plt.subplot(212, sharex=ax1)

        # Plot BTC price in upper subplot
        ax1.plot(merged_data['datetime'], merged_data['btc_price'], color='orange')
        ax1.set_title('BTC Price vs fUSD Funding Rate')
        ax1.set_ylabel('BTC Price (USD)')
        ax1.grid(True)

        # Plot fUSD rate in lower subplot
        ax2.plot(merged_data['datetime'], merged_data['daily_rate_percent'], color='blue')
        ax2.set_ylabel('fUSD Daily Rate (%)')
        ax2.set_xlabel('Date')
        ax2.grid(True)

        plt.tight_layout()
        plt.savefig('btc_fusd_comparison.png')
        plt.show()

    # 8. Display statistics
    if merged_data is not None and not merged_data.empty:
        print("\nfUSD Funding Rate Statistics:")
        print(f"Average daily rate: {merged_data['daily_rate_percent'].mean():.4f}%")
        print(f"Maximum daily rate: {merged_data['daily_rate_percent'].max():.4f}%")
        print(f"Minimum daily rate: {merged_data['daily_rate_percent'].min():.4f}%")
        print(f"Standard deviation: {merged_data['daily_rate_percent'].std():.4f}%")

        print("\nBTC Price Statistics:")
        print(f"Average price: ${merged_data['btc_price'].mean():.2f}")
        print(f"Maximum price: ${merged_data['btc_price'].max():.2f}")
        print(f"Minimum price: ${merged_data['btc_price'].min():.2f}")
        print(f"Standard deviation: ${merged_data['btc_price'].std():.2f}")

        # Handle missing values
        merged_data = merged_data.fillna(method='ffill').fillna(method='bfill')

        # Define features to analyze
        feature_columns = [
            'daily_rate_percent',
            'avg_period',
            'funding_amount',
            'funding_amount_used',
            'usage_ratio',
            'funding_below_threshold',
            'btc_price',
            'btc_price_change_percent',
            'btc_volatility_percent',
            'btc_volume'
        ]

        # Analyze feature correlations
        print("\nAnalyzing feature correlations...")
        plot_feature_importance(merged_data, feature_columns)

    # 9. Train multi-feature LSTM model (including BTC price features)
    if merged_data is not None and len(merged_data) > 30:  # Ensure enough data for training
        print("\nStarting to train LSTM model with BTC price features...")
        seq_length = 8  # Use past 8 data points to predict the next point

        # Define features to use (including BTC-related features)
        feature_columns = [
            'daily_rate_percent',  # Target feature
            'avg_period',          # Average funding period
            'funding_amount_used', # Funding amount used
            'usage_ratio',         # Usage ratio
            'funding_below_threshold',  # Total funding below threshold
            'btc_price',           # BTC price
            'btc_price_change_percent',  # BTC price change percentage
            'btc_volatility_percent'     # BTC volatility
        ]

        target_feature = 'daily_rate_percent'

        model, scalers, X_test, y_test, X_train, y_train, feature_data = train_multifeature_lstm_model(
            merged_data,
            feature_columns,
            seq_length=seq_length,
            epochs=100,
            batch_size=32
        )

        # 10. Evaluate model
        print("\nModel Evaluation:")
        metrics = evaluate_multifeature_model(model, X_test, y_test, scalers, target_feature)

        # 11. Plot prediction results and forecast next 7 days
        print("\nPlotting prediction results and future trends...")
        future_predictions = plot_multifeature_predictions(
            merged_data,
            model,
            scalers,
            feature_columns,
            target_feature,
            seq_length,
            future_days=7
        )

        # 12. Save prediction results
        today = datetime.now().strftime('%Y%m%d')
        future_predictions.to_csv(f"fusd_btc_predictions_{today}.csv", index=False)
        print(f"Prediction results saved to fusd_btc_predictions_{today}.csv")

        # 13. Save model (optional)
        model.save(f"fusd_btc_lstm_model_{today}.h5")
        print(f"Model saved to fusd_btc_lstm_model_{today}.h5")

        # 14. Compare model performance before and after adding BTC features
        print("\nPerforming additional experiment: Comparing model performance with and without BTC features")

        # Model without BTC features
        non_btc_features = [
            'daily_rate_percent',
            'avg_period',
            'funding_amount_used',
            'usage_ratio',
            'funding_below_threshold'
        ]

        print("\nTraining model without BTC features...")
        non_btc_model, non_btc_scalers, non_btc_X_test, non_btc_y_test, _, _, _ = train_multifeature_lstm_model(
            merged_data,
            non_btc_features,
            seq_length=seq_length,
            epochs=100,
            batch_size=32
        )

        print("\nEvaluation of model without BTC features:")
        non_btc_metrics = evaluate_multifeature_model(
            non_btc_model,
            non_btc_X_test,
            non_btc_y_test,
            non_btc_scalers,
            target_feature
        )

        future_predictions = plot_multifeature_predictions(
            merged_data,
            model,
            scalers,
            feature_columns,
            target_feature,
            seq_length,
            future_days=7
        )

        # Compare R² values of both models
        btc_r2 = metrics[3]
        non_btc_r2 = non_btc_metrics[3]

        print("\nModel Performance Comparison:")
        print(f"Model with BTC features R²: {btc_r2:.6f}")
        print(f"Model without BTC features R²: {non_btc_r2:.6f}")

        if btc_r2 > non_btc_r2:
            improvement = ((btc_r2 - non_btc_r2) / non_btc_r2) * 100
            print(f"Adding BTC features improved model performance by {improvement:.2f}%")
        else:
            decrease = ((non_btc_r2 - btc_r2) / non_btc_r2) * 100
            print(f"Adding BTC features decreased model performance by {decrease:.2f}%")
    else:
        print("Insufficient data to train LSTM model, need at least 30 records")

if __name__ == "__main__":
    main()
```
